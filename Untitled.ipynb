Q1. Total Multiplications = n*k*m And Total additions = n*m*(k-1) will be needed in order to perform a matrix multiplication between a (n,k) and (k,m) matrix. Resultant matrix will be of (n x m) so, total n*m elements will be calculated where each element will need k multiplications and k-1 additions.  

```python
import numpy as np
import time

def matrix_multiplication_by_lists(m1, m2):
    # Check dimensions
    if len(m1[0]) != len(m2):
        print("Please input valid matrices!")
        return
    
    # Create a result matrix with dimensions n x m
    ans = [[0 for _ in range(len(m2[0]))] for _ in range(len(m1))]
    
    # Perform matrix multiplication
    for i in range(len(m1)):
        for j in range(len(m2[0])):
            for k in range(len(m2)):
                ans[i][j] += m1[i][k] * m2[k][j]
    
    return ans
    

def matrix_multiplication_by_numpy(m1, m2):
    return np.dot(m1, m2)


m1 = [list(range(0,100))]*10
m2 = [list(range(0,100))]*100

# Time the list of lists implementation
start = time.time()
result_by_lists = matrix_multiplication_by_lists(m1, m2)
end = time.time()
time_in_lists = end - start
print(f"Time using lists: {time_in_lists:.6f} seconds")

matrix1 = np.array(matrix1)
matrix2 = np.array(matrix2)

# Time the NumPy implementation
start = time.time()
result_by_numpy = matrix_multiplication_by_numpy(m1, m2)
end = time.time()
time_in_numpy = end - start
print(f"Time using NumPy: {time_in_numpy:.6f} seconds")
```

```python
# Time Complexity: O(n*n)

import numpy as np
import sys
import timeit

def median(arr):
    length = len(arr)//2
    latestHighest = sys.maxsize

    # When list size is even 
    if(len(arr)%2 == 0):
        median = 0
        for i in range(length+1):
            temp_max = -sys.maxsize - 1
            for j in arr:
                if(j >= temp_max and j < latestHighest):
                    temp_max = j
            latestHighest = temp_max
            if(i == length - 1 or i == length):
                median = median + latestHighest

        median = median/2
        return median

    # When list size is odd
    else:
        for i in range(length+1):
            temp_max = -sys.maxsize - 1
            for j in arr:
                if(j >= temp_max and j < latestHighest):
                    temp_max = j
            latestHighest = temp_max
        return latestHighest
    

def median_opt_helper(arr, k):

    pivot = arr[len(arr)//2]

    low = [x for x in arr if x < pivot]
    high = [x for x in arr if x > pivot]
    pivots = [x for x in arr if x == pivot]

    if k < len(low):
        return median_opt_helper(low, k)
    elif k < len(low) + len(pivots):
        return pivots[0]
    else:
        return median_opt_helper(high, k - len(low) - len(pivots))


# Time complexity: O(n) [Average case scenario]
def median_opt(arr):

    n = len(arr)
    if n % 2 == 1:
        return median_opt_helper(arr, n // 2)
    else:
        return (median_opt_helper(arr, n // 2 - 1) + median_opt_helper(arr, n // 2)) / 2

def median_numpy(arr):
    return np.median(arr)


arr = np.random.uniform(1,100, size=(1,250))[0]
%timeit -r10 -n20 median(arr)
%timeit -r10 -n20 median_opt(arr)
%timeit -r10 -n20 median_numpy(arr)
    

# Time: median_numpy < median_opt < median for in general big inputs.
```

Q4. With respect to x : 2x*y + y^3 * cos(x) &
    With respect to y : x^2 + 3*y^2*sin(x)

```python
import jax.numpy as jnp
from jax import grad

def f(x, y):
  return x**2 * y + y**3 * jnp.sin(x)

def df_dx(x, y):
  return 2*x*y + y**3 * jnp.cos(x)

def df_dy(x, y):
  return x**2 + 3*y**2 * jnp.sin(x)

# Let's check the gradients at x = 4, y = 5
x, y = 4.0, 5.0

print(df_dx(x, y), grad(f, 0)(x, y))
print(df_dy(x, y), grad(f, 1)(x, y))

# Let's check the gradients at x = -0.5, y = -2.5
x, y = -0.5, -2.5

print(df_dx(x, y), grad(f, 0)(x, y))
print(df_dy(x, y), grad(f, 1)(x, y))
```

```python
import numpy as np
import pandas as pd

# Generate a matrix of size 20x5 containing random numbers drawn uniformly from the range of 1 to 2
matrix = np.random.uniform(1, 2, size=(20, 5))

# Create a dataframe out of the matrix and name the columns "a", "b", "c", "d", "e"
df = pd.DataFrame(matrix, columns=["a", "b", "c", "d", "e"])
print(df)

# Find the column with the highest standard deviation
column_std = df.std()
highest_std_column = column_std.idxmax()
print(f"Column with highest standard deviation: {highest_std_column}")

# Find the row with the lowest mean
row_mean = df.mean(axis=1)
lowest_mean_row = row_mean.idxmin()
print(f"Row with lowest mean: {lowest_mean_row}")

# Add a new column "f" which is the sum of the columns "a", "b", "c", "d", "e"
df["f"] = df.sum(axis=1)

# Create a column "g" which is "LT8" if the value in the column "f" is less than 8, and "GT8" otherwise
df["g"] = np.where(df["f"] > 8, "GT8", "LT8")
print(df)

# Find the number of rows in the dataframe where the value in the column "g" is "LT8"
num_lt8 = len(df[df["g"] == "LT8"])
print(f"Number of rows where 'g' value is 'LT8': {num_lt8}")

# Find the standard deviation of the column "f" for the rows where the value in the column "g" is "LT8" and "GT8" respectively
lt8_std = df[df["g"] == "LT8"]["f"].std()
gt8_std = df[df["g"] == "GT8"]["f"].std()
print(f"Standard deviation of 'f' for rows with 'g' value 'LT8': {lt8_std:.2f}")
print(f"Standard deviation of 'f' for rows with 'g' value 'GT8': {gt8_std:.2f}")
```
